```python
from sklearn.decomposition import PCA
from sklearn.decomposition import FactorAnalysis
from sklearn.preprocessing import LabelEncoder

# Perform PCA on engagement metrics
pca_engagement = PCA(n_components=1)
engagement_metrics['Overall Score PCA'] = pca_engagement.fit_transform(engagement_metrics)

# Perform PCA on training metrics
pca_training = PCA(n_components=1)
training_metrics['Overall Score PCA'] = pca_training.fit_transform(training_metrics)

# Perform PCA on turnover metrics
pca_turnover = PCA(n_components=1)
turnover_metrics['Overall Score PCA'] = pca_turnover.fit_transform(turnover_metrics)

# Perform PCA on diversity metrics
pca_diversity = PCA(n_components=1)
diversity_metrics_filled = diversity_metrics.fillna(diversity_metrics.mean())
diversity_metrics_filled['Overall Score PCA'] = pca_diversity.fit_transform(diversity_metrics_filled)

# Perform label encoding and PCA on time to adopt metrics
label_encoder = LabelEncoder()
time_to_adopt_metrics['Encoded_Column'] = label_encoder.fit_transform(time_to_adopt_metrics['How satisfied are you with the efficiency of the recruitment process when you joined the company?'])

pca_time_to_adopt = PCA(n_components=1)
time_to_adopt_metrics['Overall Score PCA'] = pca_time_to_adopt.fit_transform(time_to_adopt_metrics[['Encoded_Column']])

# Perform Factor Analysis on engagement metrics
fa_engagement = FactorAnalysis(n_components=1)
engagement_metrics['Overall Score Factor Analysis'] = fa_engagement.fit_transform(engagement_metrics)

# Perform Factor Analysis on training metrics
fa_training = FactorAnalysis(n_components=1)
training_metrics['Overall Score Factor Analysis'] = fa_training.fit_transform(training_metrics)

# Perform Factor Analysis on turnover metrics
fa_turnover = FactorAnalysis(n_components=1)
turnover_metrics['Overall Score Factor Analysis'] = fa_turnover.fit_transform(turnover_metrics)

# Perform Factor Analysis on diversity metrics
fa_diversity = FactorAnalysis(n_components=1)
diversity_metrics_filled = diversity_metrics.fillna(diversity_metrics.mean())
diversity_metrics_filled['Overall Score Factor Analysis'] = fa_diversity.fit_transform(diversity_metrics_filled)

# Perform Factor Analysis on time to adopt metrics
fa_time_to_adopt = FactorAnalysis(n_components=1)
time_to_adopt_metrics['How satisfied are you with the efficiency of the recruitment process when you joined the company?'] = label_encoder.fit_transform(time_to_adopt_metrics['How satisfied are you with the efficiency of the recruitment process when you joined the company?'])
time_to_adopt_metrics['Overall Score Factor Analysis'] = fa_time_to_adopt.fit_transform(time_to_adopt_metrics)



```

```python

# Compare results for Employee Engagement Metrics
if 'Overall Score PCA' in engagement_metrics.columns and 'Overall Score Factor Analysis' in engagement_metrics.columns:
    engagement_comparison = pd.DataFrame({
        'Overall Score PCA': engagement_metrics['Overall Score PCA'],
        'Overall Score Factor Analysis': engagement_metrics['Overall Score Factor Analysis']
    })
    print("Employee Engagement Metrics:")
    print(engagement_comparison)
else:
    print("Overall Score columns not found in Employee Engagement Metrics DataFrame.")

# Compare results for Training Metrics
if 'Overall Score PCA' in training_metrics.columns and 'Overall Score Factor Analysis' in training_metrics.columns:
    training_comparison = pd.DataFrame({
        'Overall Score PCA': training_metrics['Overall Score PCA'],
        'Overall Score Factor Analysis': training_metrics['Overall Score Factor Analysis']
    })
    print("\nTraining Metrics:")
    print(training_comparison)
else:
    print("Overall Score columns not found in Training Metrics DataFrame.")

# Compare results for Turnover Metrics
if 'Overall Score PCA' in turnover_metrics.columns and 'Overall Score Factor Analysis' in turnover_metrics.columns:
    turnover_comparison = pd.DataFrame({
        'Overall Score PCA': turnover_metrics['Overall Score PCA'],
        'Overall Score Factor Analysis': turnover_metrics['Overall Score Factor Analysis']
    })
    print("\nTurnover Metrics:")
    print(turnover_comparison)
else:
    print("Overall Score columns not found in Turnover Metrics DataFrame.")

# Compare results for Diversity Metrics
if 'Overall Score PCA' in diversity_metrics_filled.columns and 'Overall Score Factor Analysis' in diversity_metrics_filled.columns:
    diversity_comparison = pd.DataFrame({
        'Overall Score PCA': diversity_metrics_filled['Overall Score PCA'],
        'Overall Score Factor Analysis': diversity_metrics_filled['Overall Score Factor Analysis']
    })
    print("\nDiversity Metrics:")
    print(diversity_comparison)
else:
    print("Overall Score columns not found in Diversity Metrics DataFrame.")

# Compare results for Time to Adopt Metrics
if 'Overall Score PCA' in time_to_adopt_metrics.columns and 'Overall Score Factor Analysis' in time_to_adopt_metrics.columns:
    time_to_adopt_comparison = pd.DataFrame({
        'Overall Score PCA': time_to_adopt_metrics['Overall Score PCA'],
        'Overall Score Factor Analysis': time_to_adopt_metrics['Overall Score Factor Analysis']
    })
    print("\nTime to Adopt Metrics:")
    print(time_to_adopt_comparison)
else:
    print("Overall Score columns not found in Time to Adopt Metrics DataFrame.")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set(style="whitegrid")

# Define a function to plot the comparison for each metric
def plot_comparison(metric_comparison, metric_name):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=metric_comparison, x='Overall Score PCA', y='Overall Score Factor Analysis')
    plt.title(f'{metric_name} Comparison')
    plt.xlabel('Overall Score PCA')
    plt.ylabel('Overall Score Factor Analysis')
    plt.show()

# Plot Employee Engagement Metrics comparison
if 'Overall Score PCA' in engagement_metrics.columns and 'Overall Score Factor Analysis' in engagement_metrics.columns:
    plot_comparison(engagement_comparison, 'Employee Engagement Metrics')
else:
    print("Overall Score columns not found in Employee Engagement Metrics DataFrame.")

# Plot Training Metrics comparison
if 'Overall Score PCA' in training_metrics.columns and 'Overall Score Factor Analysis' in training_metrics.columns:
    plot_comparison(training_comparison, 'Training Metrics')
else:
    print("Overall Score columns not found in Training Metrics DataFrame.")

# Plot Turnover Metrics comparison
if 'Overall Score PCA' in turnover_metrics.columns and 'Overall Score Factor Analysis' in turnover_metrics.columns:
    plot_comparison(turnover_comparison, 'Turnover Metrics')
else:
    print("Overall Score columns not found in Turnover Metrics DataFrame.")

# Plot Diversity Metrics comparison
if 'Overall Score PCA' in diversity_metrics_filled.columns and 'Overall Score Factor Analysis' in diversity_metrics_filled.columns:
    plot_comparison(diversity_comparison, 'Diversity Metrics')
else:
    print("Overall Score columns not found in Diversity Metrics DataFrame.")

# Plot Time to Adopt Metrics comparison
if 'Overall Score PCA' in time_to_adopt_metrics.columns and 'Overall Score Factor Analysis' in time_to_adopt_metrics.columns:
    plot_comparison(time_to_adopt_comparison, 'Time to Adopt Metrics')
else:
    print("Overall Score columns not found in Time to Adopt Metrics DataFrame.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set(style="whitegrid")

# Define a function to plot the comparison for each metric
def plot_comparison(metric_comparison, metric_name):
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=metric_comparison, palette=['blue', 'orange'])
    plt.title(f'{metric_name} Comparison')
    plt.xlabel('Index')
    plt.ylabel('Overall Score')
    plt.legend(['PCA', 'Factor Analysis'])
    plt.show()

# Plot Employee Engagement Metrics comparison
if 'Overall Score PCA' in engagement_metrics.columns and 'Overall Score Factor Analysis' in engagement_metrics.columns:
    plot_comparison(engagement_comparison, 'Employee Engagement Metrics')
else:
    print("Overall Score columns not found in Employee Engagement Metrics DataFrame.")

# Plot Training Metrics comparison
if 'Overall Score PCA' in training_metrics.columns and 'Overall Score Factor Analysis' in training_metrics.columns:
    plot_comparison(training_comparison, 'Training Metrics')
else:
    print("Overall Score columns not found in Training Metrics DataFrame.")

# Plot Turnover Metrics comparison
if 'Overall Score PCA' in turnover_metrics.columns and 'Overall Score Factor Analysis' in turnover_metrics.columns:
    plot_comparison(turnover_comparison, 'Turnover Metrics')
else:
    print("Overall Score columns not found in Turnover Metrics DataFrame.")

# Plot Diversity Metrics comparison
if 'Overall Score PCA' in diversity_metrics_filled.columns and 'Overall Score Factor Analysis' in diversity_metrics_filled.columns:
    plot_comparison(diversity_comparison, 'Diversity Metrics')
else:
    print("Overall Score columns not found in Diversity Metrics DataFrame.")

# Plot Time to Adopt Metrics comparison
if 'Overall Score PCA' in time_to_adopt_metrics.columns and 'Overall Score Factor Analysis' in time_to_adopt_metrics.columns:
    plot_comparison(time_to_adopt_comparison, 'Time to Adopt Metrics')
else:
    print("Overall Score columns not found in Time to Adopt Metrics DataFrame.")
```
